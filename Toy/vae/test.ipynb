```python
import numpy as np

# 주어진 데이터
data_x = np.array([i for i in range(101)] + [0 for i in range(101)])
data_y = np.array([0 for i in range(101)] + [i for i in range(101)])

# 피어슨 상관계수 계산
correlation_coefficient = np.corrcoef(data_x, data_y)[0, 1]

print("피어슨 상관계수:", correlation_coefficient)
```

```python
import matplotlib.pyplot as plt


plt.scatter(data_x, data_y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([1,1.2,1.3,2,0.9])
x2 = np.array([1,2,1.5,1.4, 10.0,11.1,8.4])
```

```python
m = np.arange(-20,20,0.1)
```

```python
def variance(x,mean):
    N = len(x)
    return((mean - x)**2).sum()/N
```

```python
plt.scatter(x1, np.array([variance(x1, mean) for mean in x1]))
```

```python
plt.scatter(x2, np.array([variance(x2, mean) for mean in x2]))
```
